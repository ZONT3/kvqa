```python
from pathlib import Path
from matplotlib import pyplot as plt
import pandas as pd

history_path = Path('../checkpoints/history-01.json')
history = pd.read_json(history_path)
history.set_index('epoch', inplace=True)
history.head()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(history['score'], label='train')
plt.plot(history['val_score'], label='val')
plt.title('Score')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend();
```

```python
plt.figure(figsize=(12, 5))
plt.plot(history['loss'], label='train')
plt.plot(history['val_loss'], label='val')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.legend();
```
